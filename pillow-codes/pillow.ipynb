```python
import os
```

```python
#resize process for multiple fotos
folder_path = '../images'
paths = os.listdir(folder_path)
print(paths)
print('---')
print(folder_path)
```

```python
for path in paths:
    if path.endswith('.jpg'):
        img = Image.open(os.path.join(folder_path, path)) # path klasörün içindeki resimin isimi olduğu için open ile fotoğrafo açmak için folder_path i de almamız lazım
        resie_image = img.copy()
        resie_image.thumbnail((300,300))
        resie_image = resie_image.transpose(Image.FLIP_LEFT_RIGHT)
        resie_image.save(f'../resize_images/{path}')
```
